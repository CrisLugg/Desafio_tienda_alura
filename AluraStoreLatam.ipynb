### Importación de datos



```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda.info()
```

#1. Análisis de facturación



```python
tiendas= ["tienda", "tienda2", "tienda3", "tienda4"]
```

```python
Facturacion_tienda = sum(tienda["Precio"].values).astype(int)
Facturacion_tienda2 = sum(tienda2["Precio"].values).astype(int)
Facturacion_tienda3 = sum(tienda3["Precio"].values).astype(int)
Facturacion_tienda4 = sum(tienda4["Precio"].values).astype(int)
```

```python
print(Facturacion_tienda, Facturacion_tienda2, Facturacion_tienda3, Facturacion_tienda4)
```

```python
Facturacion_por_tienda = [Facturacion_tienda, Facturacion_tienda2, Facturacion_tienda3, Facturacion_tienda4]
```

```python
plt.bar(x=tiendas, height=Facturacion_por_tienda)
plt.show()
```

# 3. Calificación promedio de la tienda


```python
calificacion_tienda =  tienda['Calificación'].mean()
calificacion_tienda2 =  tienda2['Calificación'].mean()
calificacion_tienda3 =  tienda3['Calificación'].mean()
calificacion_tienda4 =  tienda4['Calificación'].mean()
```

```python
Calificacion_tiendas = [calificacion_tienda, calificacion_tienda2, calificacion_tienda3, calificacion_tienda4]
```

# 5. Envío promedio por tienda

```python
Envio_prom_por_tienda = tienda["Costo de envío"].mean().round(1)
Envio_prom_por_tienda2 = tienda2["Costo de envío"].mean().round(1)
Envio_prom_por_tienda3 = tienda3["Costo de envío"].mean().round(1)
Envio_prom_por_tienda4 = tienda4["Costo de envío"].mean().round(1)
```

```python
print(Envio_prom_por_tienda, Envio_prom_por_tienda2, Envio_prom_por_tienda3, Envio_prom_por_tienda4)
```

```python
Envio_prom_por_tiendas = [Envio_prom_por_tienda, Envio_prom_por_tienda2, Envio_prom_por_tienda3, Envio_prom_por_tienda4]
```

```python
colores = ['red', 'blue', 'green', 'orange']


plt.figure(figsize=(8,6))
for tienda, envio, calificacion, color in zip(tiendas, Envio_prom_por_tiendas, Calificacion_tiendas, colores):
    plt.scatter(envio, calificacion, color=color, label=tienda, s=100)  # s=100 para puntos más grandes
    plt.text(envio+0.1, calificacion, tienda)  # etiqueta cerca del punto

plt.xlabel('Costo de envío')
plt.ylabel('Calificación')
plt.title('Envío vs Calificación por Tienda')
plt.legend()
plt.show()
```
